```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests 
```

```python
tips = sns.load_dataset("tips")
print (tips)
```

```python
sns.set_style('white')

sns.lineplot(data = tips,
             x='total_bill',
             y= 'tip')
```

```python
tipgraph = sns.lineplot(data = tips,
                        x='total_bill',
                        y='tip')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amount ($)')

```

```python
fig, ax = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='hotpink',
    linestyle='--',
    linewidth=3,
    marker='o',  # Use a valid marker
    markerfacecolor='indigo',
    ax=ax
)
```

```python
tipgraph = sns.scatterplot(
    data=tips, 
    x='total_bill',
    y='tip', 
    style='time', 
    hue='day',
    palette=['purple', 'yellowgreen', 'orange', 'blue']  # Include enough colors for each unique 'day'
)

tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'  # Corrected the typo
)
```

```python
sns.pairplot(data = tips, 
             hue = 'day')
```

```python
daysplot = sns.relplot(data=tips,
                       x="total_bill",
                       y="tip",
                       hue="sex",
                       col="day",
                       kind="scatter",
                       col_wrap=2)
```

```python
import plotly.graph_objects as go 

x1 = np.array(["luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
```

```python
graph.update_layout(title="Pirate Scores", 
                    xaxis_title="Pirates",
                    yaxis_title="Score")
graph.show()
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x=x1, 
    y=y1, 
    mode='markers',
    marker=dict(
        size=10,  # Numeric value for size
        color='hotpink',  # Color of the markers
        opacity=1,
        line=dict(
            width=2,  # Numeric value for width
            color='purple'
        )
    )
))

graph.update_layout(
    title='Interactive Pirate Plot',
    xaxis_title='Pirates',
    yaxis_title='Scores',
    width=500, 
    height=500
)

graph.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines='skip')

# Combine all quotes into a single string
text = " ".join(quote for quote in df['quote'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
text = "".join(each for each in df.quote)

# Generate the word cloud
wordcloud = WordCloud(background_color='white', colormap='inferno').generate(text)

# Create a plot
fig, ax = plt.subplots(figsize=(7, 3))

# Display the word cloud
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")  # Turn off the axis

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_unweighted
```

```python
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", 'red'),
                           alpha=0.5)
plt.show()
```

```python
diagram.get_label_by_id("10").set_text("n".join(A - B))

diagram.get_label_by_id("11").set_text("n".join(A & B))

diagram.get_label_by_id("01").set_text("n".join(B - A))

plt.show()
```
