```python
# Custom font settings
font1 = {'family' : 'sans-serif', 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'monospace', 'color':'green', 'size' :14}
```

```python
# Creating the plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
# Adding title and labels with custom fonts
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel ('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
# Adjusting the layout to ensure everything fits
fig.tight_layout()
```

```python
# Creating the plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
# Adding title and labels with custom fonts
ax.set_title('total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
# Adjusting the layout to ensure everything fits
fig.tight_layout()
```

```python
# Creating the plot
fig, ax = plt.subplots(figsize=(5, 3))

# Plotting the data with the specified marker and line styles
ax.plot(
    x, y, 
    marker='*',
    markersize=12,
    color='#7425b9',
    linestyle='--',
    linewidth=2,
    markeredgecolor='#fa9359',
    markerfacecolor='#fa9359'
)

# Adding a grid on the y-axis with customizations
ax.grid(axis='y', color='blue', linestyle='-.', linewidth=2)

# Adding labels and title
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Customized Line Plot with Grid')

# Displaying the plot
plt.show()
```
