```python
import pandas as pd

file_path = r'C:\UFT data science program\DSI_1\visualization\02_activities\assignments\gdp_goods.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Inspect the first few rows of the dataframe
print(df.head())
```

```python
# Extract relevant columns
df_latest = df[['GDP by Goods Producing Industries, Ontario, 2007-2023', 'Unnamed: 18']]

# Rename columns for clarity
df_latest.columns = ['Industry', 'GDP_2023']

# Drop any rows where GDP is not available
df_latest = df_latest.dropna()

# Inspect the cleaned dataframe
print(df_latest.head())
```

```python
# Set the industry as the index
df_latest.set_index('Industry', inplace=True)

# Prepare data for the pie chart
labels = df_latest.index
sizes = df_latest['GDP_2023']
```

```python
import matplotlib.pyplot as plt

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(labels))))

# Add a title
plt.title('GDP Distribution by Goods Producing Industries (2023)')

# Show the pie chart
plt.show()
```
